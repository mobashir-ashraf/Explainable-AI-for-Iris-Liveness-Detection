## 1. Literature Review: Iris Biometrics & Presentation Attack Detection (PAD)
Iris recognition is highly accurate due to the distinct and permanent micro-textures found in the iris. However, security loops are vulnerable to several Presentation Attacks (PAs):
* **Printed Eye Images:** High-resolution 2D printouts designed to fool basic camera sensors.
* **Replay Attacks:** Playing back recorded video sequences of a live eye on digital screens.
* **Synthetic Iris Patterns:** Artificially generated mathematical textures created to impersonate human eyes.
* **Textured Contact Lenses:** Cosmetic lenses printed with fake topologies to alter or obscure genuine traits.

### The Black-Box Challenge
Traditional systems perform binary classification (Live vs. Spoof) with high accuracy but offer no structural explanations. They are "black-box" models. This project overcomes this by using Explainable AI (XAI) to map out decision-making regions and decomposing prediction confidence into distinct biometric factors.

## 2. Dataset Understanding Document
To evaluate the framework's robustness across different attack vectors, we explore the following standard public datasets:
1. **Warsaw Iris Dataset:** Specialized in evaluating high-resolution print attacks.
2. **NDCLD'15 Dataset:** The industry standard for cosmetic and textured contact lens detection.
3. **CASIA-Iris Dataset:** Comprehensive collection spanning synthetic iris patterns and printed variants.
4. **EyeDentify Dataset:** Composed of eye video sequences, perfect for evaluating temporal behaviors.
